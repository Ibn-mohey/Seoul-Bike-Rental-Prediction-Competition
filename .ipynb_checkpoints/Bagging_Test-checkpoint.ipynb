```python
from sklearn.feature_selection import RFECV
features = df_prep.columns.to_list()
features.remove('y')
features.remove('Date')
features.remove('Date_tmp')
features.remove('Prev Month y')

# define dataset
train = df_prep.copy()
train["y"] = np.log(df_prep["y"]+0.00001) ##--- someone explain why this works better than +1
#train["y"] = df_prep["y"]

X1 = train[features]
y1 = train["y"]
X_train, X_test, y_train, y_test =\
train_test_split(X1, y1, test_size = 0.2, random_state=42,shuffle=True,)
GBR = GradientBoostingRegressor(random_state =0,loss='ls',
    n_estimators=2500,subsample=0.8,min_samples_leaf = 0.0001
    ,max_features=0.9,learning_rate=0.04
    ,verbose =0, criterion='friedman_mse')
```

```python
RFECV = RFECV(estimator = GBR, step=1, min_features_to_select=20
                                , cv=5, scoring= scoring
                                ,n_jobs=2)
RFECV.fit(X_train , y_train)
```

```python
y_pred_trainGBR = RFECV.predict(X_train)  
y_pred_testGBR = RFECV.predict(X_test)  

rmsle_trainGBR =  rmsle(y_train, y_pred_trainGBR, True)
rmsle_testGBR  = rmsle(y_test, y_pred_testGBR, True)
print('(Train)  GBR  RMSLE:',rmsle_trainGBR )
print('(Test) GBR RMSLE:',rmsle_testGBR  )
print('---------------------------------' )
print('(Train) Prev RMSLE:', prev_trainGBR)
print('(Test) Prev RMSLE:', prev_testGBR)
prev_trainGBR = rmsle_trainGBR
prev_testGBR = rmsle_testGBR 
print('---------------------------------' )
print("Training Score:", r2_score(y_train, y_pred_trainGBR) * 100,' %')
print("Test Score:", r2_score(y_test, y_pred_testGBR) * 100,' %')
```

```python
# evaluate bagging ensemble for regression
#2500 

# define dataset
train = df_prep.copy()
train["y"] = np.log(df_prep["y"]+0.00001) ##--- someone explain why this works better than +1
#train["y"] = df_prep["y"]

X1 = train[selected_features]
y1 = train["y"]
X_train, X_test, y_train, y_test =\
train_test_split(X1, y1, test_size = 0.2, random_state=42,shuffle=True,)
# define the model
GBR = GradientBoostingRegressor(random_state =0,loss='ls',
    n_estimators=2500,subsample=0.8,min_samples_leaf = 0.0001
    ,max_features=0.9,learning_rate=0.04
    ,verbose =0, criterion='friedman_mse')
# BaggingRegressor(base_estimator=None, n_estimators=10, *, max_samples=1.0, max_features=1.0
#                  , bootstrap=True, bootstrap_features=False, oob_score=False, warm_start=False
#                  , n_jobs=None, random_state=None, verbose=0)
BaggModel = BaggingRegressor(base_estimator = GBR , n_estimators =10 ,  bootstrap=True 
                            , bootstrap_features=True ,n_jobs=3, random_state=911 )
BaggModel.fit(X_train , y_train)
# evaluate the model
y_pred_trainBaggingRegressor = BaggModel.predict(X_train)  
y_pred_testBaggingRegressor = BaggModel.predict(X_test)  

rmsle_trainBaggingRegressor =  rmsle(y_train, y_pred_trainBaggingRegressor, True)
rmsle_testBaggingRegressor  = rmsle(y_test, y_pred_testBaggingRegressor, True)
```
